### Qual deve ser a nova cidade brasileira a ter uma Starbucks? 

```python
dados_brasil_limpo['%PEA'] = dados_brasil_limpo['População economicamente ativa de 18 anos ou mais de idade 2010']/dados_brasil_limpo['População total 2010']
```

```python
dados_brasil_limpo.drop('População economicamente ativa de 18 anos ou mais de idade 2010', axis=1, inplace=True)
```

```python
dados_brasil_limpo.columns.values[0:4] =["Municípios", "População","Serviços","Renda"]
```

```python
dados_brasil_limpo
```

```python
#Normalizando as Variáveis
dados_brasil_limpo['normalized_pop'] = (dados_brasil_limpo['População']-dados_brasil_limpo['População'].min())/(dados_brasil_limpo['População'].max()-dados_brasil_limpo['População'].min())
dados_brasil_limpo['normalized_serv'] = (dados_brasil_limpo['Serviços']-dados_brasil_limpo['Serviços'].min())/(dados_brasil_limpo['Serviços'].max()-dados_brasil_limpo['Serviços'].min())
dados_brasil_limpo['normalized_rend'] = (dados_brasil_limpo['Renda']-dados_brasil_limpo['Renda'].min())/(dados_brasil_limpo['Renda'].max()-dados_brasil_limpo['Renda'].min())
dados_brasil_limpo['normalized_pea'] = (dados_brasil_limpo['%PEA']-dados_brasil_limpo['%PEA'].min())/(dados_brasil_limpo['%PEA'].max()-dados_brasil_limpo['%PEA'].min())
```

```python
dados_brasil_limpo
```

```python
peso_pop = 1
peso_serv = 1
peso_rend = 10
peso_pea = 100
dados_brasil_limpo['score'] = (peso_pop*dados_brasil_limpo['normalized_pop'] + peso_serv*dados_brasil_limpo['normalized_serv'] + peso_rend*dados_brasil_limpo['normalized_rend'] + peso_pea*dados_brasil_limpo['normalized_pea'])/(peso_pop+peso_serv+peso_rend+peso_pea)
```

```python
dados_brasil_limpo.sort_values(by='score', ascending = False).head(50)
```
